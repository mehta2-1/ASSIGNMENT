Q1-Explain the difference between Relative , Absolute and Mixed Cell Referencing.

ANS- (1)RELATIVE REFERNCING:
      
      -The cell address automatically gets changed when the formula is copied.

      -EXAMPLE:cell address "A1" gets changed into "A2" when the formula is dragged down .

      (2)ABSOLUTE REFERENCING:
      
       -here we lock the desired cell address which we want to be fixed throughout,while using the same formula.

       -EXAMPLE:$A$1 , here using "$" we have fixed the column as well the row of cell address A1 .

       (3)MIXED REFERENCING:
         
         -here we can either fix/lock the column or row of a cell adress in a formula according tomthe requirement.

         -EXAMPLE: $A1 / A$1 both falls under mixed referencing.
